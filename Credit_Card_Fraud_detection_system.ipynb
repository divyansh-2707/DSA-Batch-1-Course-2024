```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
#loading credit card dataset
df=pd.read_csv('/content/drive/MyDrive/creditcard.csv')
```

```python
#here dropna function is used for removing null values from the dataset
df=df.dropna(how="any")


```

```python
df.isnull().sum()
```

```python
# a distribution of legit and fraudlent transcation
df['Class'].value_counts()
```

```python
# 0->Normal Transcations
# 1->Fraudlent Transcations
# As we can see there is a huge difference between the normal and fraudlent data counts, this shows that data is very much imbalanced
# Now we have to balanced it first
normal_data=df[df.Class==0]
fraud_data=df[df.Class==1]
print(normal_data.shape)
print(fraud_data.shape)
```

```python
#statistical measures of these two dataset
normal_data.describe()
```

```python
fraud_data.describe()
```

Undersampling -> To handle the unbalanced data


```python
#As we can see the fraudlent data is lesser in number so we try to equal the normal transcation with that number for making it balance.
sample_data=normal_data.sample(n=52)
```

```python
sample_data.shape
```

```python
sample_data.describe()
```

```python
sample_data.info()
```

```python
sample_data['Class'].value_counts()
```

```python
#Concatenation of the two datasets
new_dataset=pd.concat([sample_data,fraud_data],axis=0)
```

```python
new_dataset.shape
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.info()
```

```python
new_dataset.describe()
```

```python
new_dataset.groupby('Class').mean()
```

```python
new_dataset.tail()
```

**Splitting the data into testing and training data**



```python
X=new_dataset.drop(columns='Class',axis=1)
Y=new_dataset['Class']
```

```python
#spliting the training and testing data
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,test_size=0.3,stratify=Y,random_state=3)
```

```python
print(X_train.shape,X_test.shape,Y_train.shape,Y_test.shape)
```

Model training

```python
#Logistic Regression ->used in binary classification
model=LogisticRegression()
```

```python
# training the model with model with Training Data
model.fit(X_train,Y_train)
```

```python
#Evaluation of the model based on Accuracy Score
```

```python
#Accuracy on training data
X_train_predict=model.predict(X_train)
training_data_accuracy=accuracy_score(X_train_predict,Y_train)
```

```python
print("Accuracy of training data-> ",training_data_accuracy*100,"%")
```

```python
#Accuracy data of testing data
X_testing_data=model.predict(X_test)
testing_data_accuracy=accuracy_score(X_testing_data,Y_test)
```

```python
print("Accuracy of Testing data ->",testing_data_accuracy * 100,"%")
```
